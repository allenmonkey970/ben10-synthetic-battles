```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/ben10_master.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df[['alien1_strength_level', 'alien2_strength_level']], ax=axes[0])
axes[0].set_title('Strength Level Distribution')
sns.boxplot(data=df[['alien1_speed_level', 'alien2_speed_level']], ax=axes[1])
axes[1].set_title('Speed Level Distribution')
sns.boxplot(data=df[['alien1_intelligence', 'alien2_intelligence']], ax=axes[2])
axes[2].set_title('Intelligence Level Distribution')
plt.show()
```

```python
top10_winners = df['winner'].value_counts().head(10).index

plt.figure(figsize=(10, 6))
sns.countplot(
    y='winner',
    data=df[df['winner'].isin(top10_winners)],
    order=top10_winners
)
plt.title('Top 10 Winners by Number of Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Winner')
plt.show()
```

```python
sns.pairplot(df, vars=[
    'alien1_strength_level', 'alien1_speed_level', 'alien1_intelligence',
    'alien2_strength_level', 'alien2_speed_level', 'alien2_intelligence'
], hue='winner', height=2.5)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
df['alien1_home_planet'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Alien Home Planets (Alien 1)')
plt.xlabel('Home Planet')
plt.ylabel('Count')
plt.show()
```
